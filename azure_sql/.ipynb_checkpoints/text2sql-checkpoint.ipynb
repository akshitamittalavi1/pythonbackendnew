### Loading tables to Azure SQL DB

```python
# function for creating table and adding csv data into table on azure database

def table_insertion(connectionString, file_path):
    """
       This function is used to read the file from its path,
       loading the data into DB.
       
       Input Parameters:
       
       connectionString
       
       file_path -> path of file where data is present.
       
    
    """
    
    
        # Connect to the Azure SQL database using the connection string
        conn = pyodbc.connect(connectionString)
        cursor = conn.cursor()

        # Split the file path by the dot (.) to get the filename and extension
        sql_table_name, extension = file_path.split(".")
        sql_table_name = file_path.split("//")[-1].split('.')[0]

        if extension == 'csv':
            excel_data = pd.read_csv(file_path)
        else:
            excel_data = pd.read_excel(file_path)

        # Convert column names to lowercase
        excel_data.columns = excel_data.columns.str.lower()

        # Get column names as a tuple
        column_names_tuple = tuple(excel_data.columns.tolist())

        # Get the column names from the DataFrame
        column_names = ", ".join(excel_data.columns)

        columns_type = {}
        for column_name, dtype in excel_data.dtypes.items():
            if dtype == 'int64':
                columns_type[column_name] = 'INTEGER'
            elif dtype == 'float64':
                columns_type[column_name] = 'FLOAT'
            elif dtype == 'bool':
                columns_type[column_name] = 'BOOLEAN'
            elif dtype == 'datetime64[ns]':
                columns_type[column_name] = 'DATETIME'
            else:
                columns_type[column_name] = 'VARCHAR(MAX)'

        col_name_type = ''
        for column_name, dtype in columns_type.items():
            if col_name_type:
                col_name_type += f', {column_name} {dtype}'
            else:
                col_name_type += f'{column_name} {dtype}'

        print(f'CREATE TABLE {sql_table_name} ({col_name_type})')
        cursor.execute(f'CREATE TABLE {sql_table_name} ({col_name_type})')
        # Iterate through rows and insert data
        for _, row in excel_data.iterrows():
            # Create a parameterized SQL INSERT statement
            placeholders = ", ".join(["?" for _ in excel_data.columns])
            sql_insert = f"INSERT INTO {sql_table_name} ({column_names}) VALUES ({placeholders})"

            # Extract the values from the row as a tuple
            values = tuple(row)
            # Execute the SQL INSERT statement with the values
            cursor.execute(sql_insert, values)
            conn.commit()

        cursor.close()
        conn.close()
```

```python
# inserting tables to DB 
table_insertion(connectionString,csv_file_path6)
```

```python
# dropping a table from DB 

def drop_table(connectionString, table_name):
    """
    This function can be used to drop tables present in 
    DB
    
    Input Parameters:
    
        connectionString
        
        table_name -> name of table to drop 
    """
        conn = pyodbc.connect(connectionString)
        cursor = conn.cursor()
        cursor.execute(f'DROP TABLE {table_name};')
        conn.commit()

        cursor.close()
        conn.close()
```

```python
# funtion to convert user text into sql

def text2sql(query:str, prompt) -> str:
    """
    This function will take user query as string
    and generate sql query using given prompt.
    
    Input Parameters:
    
    query -> user text query
    
    prompt -> Prompt used to create sql query.
    
    """
    query = query.lower()
    prompt=PromptTemplate.from_template(prompt)
    chain = LLMChain(llm=chat, prompt=prompt, verbose=False)
    result = chain({'query':query})
    return result['text']
    
```

### Method 2: Using Langchain query generation

```python
def text2sql_with_chain(query:str, chat_llm, db)-> str:
    """
    This function will take user query as string and 
    generate sql query.
    
    Input Parameters:
    
    query -> user text query
    
    chat_llm -> llm object(ChatOpenAI)
    
    db -> SQLDatabase object db
    """
    chain = create_sql_query_chain(chat, db)
    response = chain.invoke({"question":query})
    return response
    
```

```python
sql = text2sql_with_chain("Retrieve the average values of 'vl1,' 'vl2,' and 'vl3' for each unique 'devicetimestamp' in the 'currentvoltage' table.", chat, db)
```

### Read results from DB 

```python
# Check inserted data in DB

def read_query(sql):
    """
    This function is used to read the output 
    generated after execution of given sql query.
    
    Input Parameter:
    
    sql -> SQL query 
    
    """
    conn = pyodbc.connect(connectionString) # opening database
    cursor = conn.cursor()  # making connection
    cursor.execute(sql) # retriving rows
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    conn.close()
    
```
